```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from sklearn.datasets import load_breast_cancer
pd.set_option("display.max_columns", None)
sns.set(style = 'whitegrid', palette = 'muted')
```

```python
data = pd.DataFrame(data = load_breast_cancer()['data'], columns = load_breast_cancer()['feature_names'])
data["diagnosis"] = load_breast_cancer()["target"]
data["diagnosis"] = data["diagnosis"].replace({0: 'Malignant', 1: 'Benign'})
data.head()
```

```python
X = data.drop(columns = ['diagnosis'], axis = 1)
y = data['diagnosis']
data.shape
```

```python
X.describe().T
```

```python
ax = sns.countplot(y, label = 'Count')
b, m = data['diagnosis'].value_counts()
```

### Violin Plot

```python
X_std = (X - X.mean()) / X.std()
data_first10 = pd.concat([y, X_std.iloc[:, 0:10]], axis = 1)
data_first10 = pd.melt(data_first10, id_vars = "diagnosis",
                      var_name = "features",
                      value_name = "value")

plt.figure(figsize = (10, 6))
sns.violinplot(x = "features", y = "value", hue = "diagnosis", data = data_first10, split = True, inner = 'quart')
plt.xticks(rotation = 90)
plt.show()
```

* To create classification model mean radius, mean texture appears to be good feature whereas mean fractal dimension would be a poor choice.
* Violin plot shows kernel density.

### Box Plot

```python
plt.figure(figsize = (10, 6))
sns.boxplot(x = "features", y = "value", hue = "diagnosis", data = data_first10)
plt.xticks(rotation = 90)
plt.show()
```

* Box plot is good when we try to find outliers in data

### Joint Plot for Feature Comparision

```python
sns.jointplot(x = X.loc[:, 'mean area'],
             y = X.loc[:, 'mean radius'],
             kind = 'reg')

plt.show()
```

### Swarm Plot

```python
plt.figure(figsize = (10, 6))
sns.stripplot(x = "features", y = "value", hue = "diagnosis", data = data_first10)
plt.xticks(rotation = 90)
plt.show()
```

In terms of class seperatbility mean area appears to be best feature.
